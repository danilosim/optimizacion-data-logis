```python
import folium
import csv
from collections import defaultdict
```

```python
with open("optimizer_result.csv", "r", encoding = "ISO-8859-1") as f:
    reader = csv.DictReader(f)
    trip_list = list(reader)

trips_by_unidad = defaultdict(list)

for trip in trip_list:
    trips_by_unidad[trip['unidad']].append(trip)
trips_by_unidad = dict(trips_by_unidad)
```

```python
m = folium.Map(location=[-38.4161, -63.6167], zoom_start=4)

longest_trip = max(trips_by_unidad.values(), key=lambda x: len(x))

for trip in longest_trip:
       loc = [
              (float(trip['origin_lat']), float(trip['origin_long'])),
              (float(trip['dest_lat']), float(trip['dest_long'])),
       ]
       folium.PolyLine(loc, color='red', weight=1, opacity=1).add_to(m)
       folium.Marker(loc[0]).add_to(m)
       folium.Marker(loc[1]).add_to(m)
m
```
